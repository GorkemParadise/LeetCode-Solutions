```python
##Google Question
def leastBricks(self, wall) -> int:
        myHashMap = {0:0}
        
        for row in wall:
            gapCount = 0
            for brick in row[:-1]:
                gapCount += brick
                myHashMap[gapCount] = 1 + myHashMap.get(gapCount,0)
                
        maximumGapNumber = max(myHashMap.values())
        rowNumber = len(wall)
        return rowNumber - maximumGapNumber
```
