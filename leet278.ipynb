```python
# isBadVersion(version) fonksiyonu zaten hazır (LeetCode sağlıyor)

class Solution:
    def firstBadVersion(self, n: int) -> int:
        left, right = 1, n

        while left < right:
            mid = left + (right - left) // 2  # orta versiyon
            if isBadVersion(mid):
                right = mid  # ilk kötü burda ya da daha solda
            else:
                left = mid + 1  # kötü versiyon daha ileride
        
        return left  # veya right, fark etmez çünkü eşit olacak
```
