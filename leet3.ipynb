```python
#1) O(n*n)
def lengthOfLongestSubstring(self,s):
    max_len = 0
    for i in range(len(s)):
        substring = ""
        for j in range(i, len(s)):
            if s[j] in substring:
                break
            substring += s[j]
        max_len = max(max_len, len(substring))
    return max_len
```

```python
#2) Hash Map
def lengthOfLongestSubstring(self,s):
    start = 0
    long = 0
    seen = {}
    for i, ch in enumerate(s):
        if ch in seen and start <= seen[ch]:
            start = seen[ch] + 1
        else:
            long = max(long, i-start+1)
        seen[ch] = i
    return long
```
