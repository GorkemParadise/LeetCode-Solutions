```python
def spiralOrder(self, matrix):
    matrix_spiral = []
    if not matrix:
        return matrix_spiral
    top, bottom = 0, len(matrix) - 1
    left, right = 0, len(matrix[0]) - 1
    while top <= bottom and left <= right:
        for i in range(left, right + 1):
            matrix_spiral.append(matrix[top][i])
        top += 1
        for i in range(top, bottom + 1):
            matrix_spiral.append(matrix[i][right])
        right -= 1
        if top <= bottom:
            for i in range(right, left - 1, -1):
                matrix_spiral.append(matrix[bottom][i])
            bottom -= 1
        if left <= right:
            for i in range(bottom, top - 1, -1):
                matrix_spiral.append(matrix[i][left])
            left += 1
    return matrix_spiral
# Example usage:
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
spiralOrder(None, matrix)  # Output: [1, 2, 3, 6, 9, 8, 7, 4, 5]
```
