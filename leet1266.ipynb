```python
def minTimeToVisitAllPoints(self, points):
    total = 0
    x1, y1 = points.pop()
    while points:
        x2, y2 = points.pop()
        total += max(abs(y2-y1), abs(x2-x1))
        x1, y1 = x2, y2
    return total
```

```python
points = [[1,1],[3,4],[-1,0]]
len(points)
```

```python
def deneme(self, points):
    t = 0
    for i in range(len(points) - 1):
        x1, y1 = points[i]
        x2, y2 = points[i + 1]
        t += max(abs(x1 - x2), abs(y1 - y2))
    return t
```

```python
deneme(None, points)
```
