```python
nums = 2, 2, 1, 1, 3

def solution(nums):
    return 2*sum(set(nums)) - sum(nums)
print(solution(nums))
```

```python
nums = [1, 1, 2, 2, 3]

def singleNumber(nums):
    return 2*sum(set(nums)) - sum(nums)
    
print(singleNumber(nums))
```

```python
nums = [1, 1, 2, 2, 3]

def single(nums):
    count = {}

    for x in nums:         
        if x in count:
            count[x] += 1
        else:
            count[x] = 1

    for key in count:     
        if count[key] == 1:
            return key
        
print(single(nums))
```
